````python
mean_new= np.mean(new_design_data)

old_mean= np.mean(old_design_data)

old_mean , mean_new
````

````python
z_cal= (mean_new - old_mean) / (np.std(old_design_data) / np.sqrt(len(old_design_data)))
z_cal
````
